**Q0.**
1. What is the difference between regression and classification?

They're both types of supervised learning techniques used in machine learning, but how they work is different. With regression, we are looking at some sort of continuous way to predict a given value, whereas classification is more like putting some object into one of many designated buckets depending on the category we predict for it. Regression could be used to predict weather temperature, or the price of something. Classification is more for things like what color is this, what season is this photo taken place, etc.

3. What is a confusion table? What does it help us understand about a model's performance?

It helps us understand the accuracy of a classification model. It is essentially a matrix  that shows the count of true/false real/false positive reports for a given set of data, showing how the model performs for each particular choice. Generally, either the row or column will represent the real values and the other will represent the reported, so we can see how well it did and in what ways the model performed incorrectly.

4. What does the SSE quantify about a particular model?

The sum of squared errors is taking the sum of all our squared differences between the actual value and the reported value, and by adding them up we get a metric to see what the overall performance was for our (generally) regression. If we had an SSE of 0, this would mean our model is perfect, as the difference between each thing and what it actually was, was 0 for all reported values.

5. What are overfitting and underfitting?

Overfitting and underfitting are both negative characteristics of a trained model for some data set, but they differ in why they are a bad attribute. With overfitting, we have trained our model too well in the context of the data set, to the point where it actually performs worse for data not in our data set. Generally, this happens when our model is trained so much it starts to see patterns that aren't actually there - such as overvaluing noise in our data that isn't actually relevant. Underfitting is a little more obvious; when our model is not trained enough it simply isn't able to identify the patterns we are hoping it can for a general data set given the data that we have provided it.

6. Why does splitting the data into training and testing sets, and choosing $k$ by evaluating accuracy or SSE on the test set, improve model performance?

Having a data set which the model is not directly trained on, but instead of as a way of gauging performance, allows us to build a generalized model that we believe will perform well in any situation we provide it. In other words, it helps us avoid overfitting. If we test our model on data that is not from the training set, we are more likely to see performance metrics that are more akin to the reality of our model than that of what we would see when comparing it to data choices we've already trained on. Our SSE for tests given via a testing set are likely to be more accurate to our model's genuine capabilities.

7. With classification, we can report a class label as a prediction or a probability distribution over class labels. Please explain the strengths and weaknesses of each approach.

Class labels are more simplistic, which means they are easier to apply, easier to understand, and require less computational power overall. Probability distributions are the opposite in all those same regards; they are more technically complicated to understand, they require more space and time to compute, and they are tougher to implement. Class labels should be used for situations where nuance is not a big factor in the way we wish to report our data.

For each value of $k$, compute the mean squared error and print a scatterplot showing the test value plotted against the predicted value. What patterns do you notice as you increase $k$?
5. Determine the optimal $k$ for these data.
6. Describe what happened in the plots of predicted versus actual prices as $k$ varied, taking your answer into part 6 into account. (Hint: Use the words "underfitting" and "overfitting".)

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
import matplotlib.pyplot as plt

# Part 1
df = pd.read_csv('data/USA_cars_datasets.csv')
df.describe()
```

```python
df.head()
```

```python
df = df.loc[:,['price','year','mileage']]
```

```python
# Part 2
normalized_df=(df-df.min())/(df.max()-df.min())
normalized_df.describe()
```

```python
# Part 3
price = df['price']
other = df.drop('price',axis=1)
X_train, X_test, y_train, y_test = train_test_split(other, price, test_size=.2, random_state=100)
```

```python
# Part 4 
k = 3
model = KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train)
prediction = model.predict(X_test)
sq_sum = np.sum( (y_test-prediction)**2)
plot, axes = plt.subplots()
plt.scatter(y_test, prediction)
plt.title('k: '+str(k)+', SSE: '+str(sq_sum))
plt.show()
```

```python
k = 10
model = KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train)
prediction = model.predict(X_test)
sq_sum = np.sum( (y_test-prediction)**2)
plot, axes = plt.subplots()
plt.scatter(y_test, prediction)
plt.title('k: '+str(k)+', SSE: '+str(sq_sum))
plt.show()
```

```python
k = 25
model = KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train)
prediction = model.predict(X_test)
sq_sum = np.sum( (y_test-prediction)**2)
plot, axes = plt.subplots()
plt.scatter(y_test, prediction)
plt.title('k: '+str(k)+', SSE: '+str(sq_sum))
plt.show()
```

```python
k = 50
model = KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train)
prediction = model.predict(X_test)
sq_sum = np.sum( (y_test-prediction)**2)
plot, axes = plt.subplots()
plt.scatter(y_test, prediction)
plt.title('k: '+str(k)+', SSE: '+str(sq_sum))
plt.show()
```

```python
k = 100
model = KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train)
prediction = model.predict(X_test)
sq_sum = np.sum( (y_test-prediction)**2)
plot, axes = plt.subplots()
plt.scatter(y_test, prediction)
plt.title('k: '+str(k)+', SSE: '+str(sq_sum))
plt.show()
```

```python
k = 300
model = KNeighborsRegressor(n_neighbors=k).fit(X_train, y_train)
prediction = model.predict(X_test)
sq_sum = np.sum( (y_test-prediction)**2)
plot, axes = plt.subplots()
plt.scatter(y_test, prediction)
plt.title('k: '+str(k)+', SSE: '+str(sq_sum))
plt.show()
```
